```python
!pip -q install --upgrade arch linearmodels transformers==4.43.3 sentencepiece \
                       statsmodels scikit-learn openpyxl nltk

import warnings, re, math, itertools
warnings.filterwarnings('ignore')

import numpy as np, pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.api import VAR, ARDL
from statsmodels.tsa.stattools import adfuller, kpss, grangercausalitytests
from statsmodels.stats.diagnostic import breaks_cusumolsresid
from sklearn.preprocessing import StandardScaler
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from nltk.stem.porter import PorterStemmer
import nltk; nltk.download('punkt', quiet=True)

try:
    import torch
    DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
except Exception:
    DEVICE = 'cpu'
print('Torch device:', DEVICE)

TXT_FILES = {
    'US': '/content/text data us.xlsx',
    'EU': '/content/text data eu.xlsx',
    'UK': '/content/text data uk.xlsx',
}
CRED_PATH  = '/content/credibility_planB.xlsx'
MACRO_PATH = '/content/chapter 4 august 2025 1st block.xlsx'

def to_monthly(s: pd.Series) -> pd.Series:
    """Safe cast to month-start (MS) from D/B/Q/M index."""
    s = s.dropna()
    if s.empty: return s
    idx = pd.DatetimeIndex(s.index)
    f = (pd.infer_freq(idx[:10]) or '').upper()
    if f.startswith(('D','B')):
        out = s.resample('MS').last()
    elif f.startswith('Q'):
        out = (s.resample('QS').last().interpolate('linear').resample('MS').ffill())
    else:
        out = s.to_period('M').to_timestamp('MS') if not idx.is_month_start.all() else s.copy()
    return out

def guess_col(df, must_have_any=(), must_have_all=()):
    """
    Επιστρέφει το 1ο column name που ταιριάζει στο pattern.
    - must_have_any: τουλάχιστον ένα substring
    - must_have_all: όλα τα substrings
    """
    for c in df.columns:
        name = str(c).lower()
        if (all(x in name for x in [x.lower() for x in must_have_all]) and
            (not must_have_any or any(x.lower() in name for x in must_have_any))):
            return c
    return None

def detect_planB_col(df, region, prefer_scaled=True):
    """
    Προσπαθεί να βρει την κατάλληλη στήλη credibility για region (US/EU/UK).
    Προτιμά scaled ('cred_sc_*EGARCH*'), αλλιώς raw ('*EGARCH*').
    """
    cols = list(df.columns)
    low  = [str(c).lower() for c in cols]
    r    = region.lower()

    sc_idx = [i for i, nm in enumerate(low) if (r in nm and 'egarch' in nm and ('cred_sc' in nm or 'scaled' in nm or 'sc_' in nm))]
    rw_idx = [i for i, nm in enumerate(low) if (r in nm and 'egarch' in nm and not ('cred_sc' in nm or 'scaled' in nm))]

    pick = None
    tag  = None
    if prefer_scaled and sc_idx:
        pick = cols[sc_idx[0]]; tag = 'scaled'
    elif rw_idx:
        pick = cols[rw_idx[0]]; tag = 'raw'
    elif sc_idx:
        pick = cols[sc_idx[0]]; tag = 'scaled'
    return pick, tag

def load_credibility_planB(path, region, prefer_scaled=True):
    """
    Διαβάζει την κατάλληλη sheet/στήλη για region από το credibility_planB.xlsx.
    Επιστρέφει Series name='credibility' (MS).
    """
    xls = pd.ExcelFile(path)
    cand = []
    for sh in xls.sheet_names:
        nm = sh.lower()
        if ('cred' in nm and 'scal' in nm) or ('raw' in nm) or ('egarch' in nm):
            cand.append(sh)
    cand = cand or xls.sheet_names

    for sh in cand:
        df0 = pd.read_excel(path, sheet_name=sh)
        c_date = guess_col(df0, must_have_any=['date', 'time', 'month'], must_have_all=[]) or 'date'
        if c_date not in df0.columns:
            continue
        c_val, tag = detect_planB_col(df0, region, prefer_scaled=prefer_scaled)
        if c_val is None:
            continue

        s = (pd.Series(df0[c_val].values,
                       index=pd.to_datetime(df0[c_date], errors='coerce'),
                       name='credibility').dropna().sort_index())
        s = to_monthly(s)
        print(f"[{region}] credibility → sheet: '{sh}', column: '{c_val}'  ({tag})")
        return s

    raise RuntimeError(f"[{region}] Δεν βρέθηκε column EGARCH/credibility στο {path}.")

REGION_ALIASES = {
    'US': ['us', 'united states', 'usa'],
    'EU': ['eu', 'euro area', 'eurozone', 'ea'],
    'UK': ['uk', 'united kingdom', 'britain', 'gb'],
}
def find_sheet_by_kind_region(xls, kind, region):
    """
    kind in {'inflation','volatility'}, region in {'US','EU','UK'}
    Γυρνά sheet που περιέχει τα κατάλληλα tokens.
    """
    k = kind.lower()
    r_alias = REGION_ALIASES.get(region, [region.lower()])
    for sh in xls.sheet_names:
        nm = sh.lower()
        if k[:4] in nm:
            if any(a in nm for a in r_alias):
                return sh
    for sh in xls.sheet_names:
        if k[:4] in sh.lower(): return sh
    return None

def load_macro_series(path, region):
    """
    Διαβάζει inflation & volatility για region από chapter 4 workbook.
    Υποθέτει ότι η 1η στήλη είναι ημερομηνία, η 2η η τιμή (όπως στο setup σου).
    Επιστρέφει (inf_gap, vol_SR), αμφότερα MS.
    """
    xls = pd.ExcelFile(path)
    sh_inf = find_sheet_by_kind_region(xls, 'inflation', region)
    sh_vol = find_sheet_by_kind_region(xls, 'volatility', region)
    assert sh_inf is not None, f"[{region}] Δεν βρέθηκε sheet για Inflation."
    assert sh_vol is not None, f"[{region}] Δεν βρέθηκε sheet για Volatility."

    df_inf = pd.read_excel(path, sheet_name=sh_inf, usecols=[0,1], names=['Date','INF'])
    df_vol = pd.read_excel(path, sheet_name=sh_vol, usecols=[0,1], names=['Date','VOL'])

    df_inf['Date'] = pd.to_datetime(df_inf['Date'], errors='coerce')
    df_vol['Date'] = pd.to_datetime(df_vol['Date'], errors='coerce')

    s_inf = to_monthly(df_inf.set_index('Date')['INF']).dropna()
    s_vol = to_monthly(df_vol.set_index('Date')['VOL']).dropna()

    inf_gap = s_inf - s_inf.mean()
    vol_SR  = s_vol
    print(f"[{region}] Macros → Inflation sheet: '{sh_inf}', Volatility sheet: '{sh_vol}'")
    return inf_gap.rename('inf_gap'), vol_SR.rename('vol_SR')

def load_texts_score_sentiment(path):
    df_txt = pd.read_excel(path)
    c_date = guess_col(df_txt, must_have_any=['date','month','time'], must_have_all=[]) or 'Date'
    c_text = guess_col(df_txt, must_have_any=['text','statement','body','content'], must_have_all=[]) or 'Text'
    assert c_date in df_txt.columns, f"Date column not found in {path}. Columns: {list(df_txt.columns)}"
    assert c_text in df_txt.columns, f"Text column not found in {path}. Columns: {list(df_txt.columns)}"

    df_txt[c_date] = pd.to_datetime(df_txt[c_date], dayfirst=True, errors='coerce')
    df_txt = (df_txt[[c_date, c_text]]
                .rename(columns={c_date:'Date', c_text:'Text'})
                .dropna(subset=['Date'])
                .set_index('Date').sort_index())
    df_txt['Text'] = df_txt['Text'].fillna('')

    global _finbert_tok, _finbert_mdl
    if '_finbert_tok' not in globals():
        print('⏳ Loading finBERT weights once …')
        _finbert_tok = AutoTokenizer.from_pretrained('ProsusAI/finbert')
        _finbert_mdl = AutoModelForSequenceClassification.from_pretrained('ProsusAI/finbert').to(DEVICE); _finbert_mdl.eval()

    def finbert_batch(texts, max_tok=512, batch=32):
        out=[]
        for i in range(0, len(texts), batch):
            sub = texts[i:i+batch]
            enc = _finbert_tok(sub, padding=True, truncation=True, max_length=max_tok, return_tensors='pt')
            enc = {k: v.to(DEVICE) for k,v in enc.items()}
            with torch.no_grad():
                logits = _finbert_mdl(**enc).logits
                prob   = torch.softmax(logits, dim=1)
                out   += (prob[:,2]-prob[:,0]).cpu().tolist()
        return out

    df_txt['sent_finbert'] = finbert_batch(df_txt['Text'].tolist())

    EXP = ['improv','foster','increas','expand','rise','higher','risen','gain','strong','acceler','faster','strength']
    CON = ['moder','slow','low','weak','subdu','lower','fall','slower','weaker','decreas','weaken','contract','soften','deceler','cool']
    AMB = ['condit','anticip','believ','risk','may','appear','conting','suggest','seem','somewhat','uncertainti','uncertain','possibl','destabil','volatil','tent','unsusu','might','alter']
    LEX = {w:'expansion' for w in EXP} | {w:'contraction' for w in CON} | {w:'ambiguity' for w in AMB}
    stem = PorterStemmer()
    def tokenize(txt): return [stem.stem(t) for t in re.findall(r"[A-Za-z']+", (txt or '').lower())]
    def hm_score(txt):
        cnt={'expansion':0,'contraction':0}
        for tok in tokenize(txt):
            k=LEX.get(tok);
            if k in cnt: cnt[k]+=1
        tot=cnt['expansion']+cnt['contraction']
        return (cnt['expansion']-cnt['contraction'])/tot if tot else np.nan
    df_txt['sent_hm'] = df_txt['Text'].apply(hm_score)

    df_m = df_txt[['sent_finbert','sent_hm']].resample('MS').mean().ffill()
    for w in (3,6,12):
        df_m[f'finbert_ema{w}m'] = df_m['sent_finbert'].ewm(span=w, adjust=False).mean()
        df_m[f'hm_ema{w}m']      = df_m['sent_hm'].ewm(span=w, adjust=False).mean()
    return df_m

def nw_lags(n):
    return max(3, int(4 * (n/100.0)**(2/9)))

def nw_ols(y, X, lags=None, add_const=True, cov='HAC', groups=None):
    X = sm.add_constant(X, has_constant='add') if add_const else X
    lags = lags or nw_lags(len(y))
    if cov == 'cluster':
        return sm.OLS(y, X, missing='drop').fit(cov_type='cluster', cov_kwds={'groups':groups})
    return sm.OLS(y, X, missing='drop').fit(cov_type='HAC', cov_kwds={'maxlags':lags})

def winsorize(s, p=0.01):
    lo, hi = s.quantile(p), s.quantile(1-p)
    return s.clip(lower=lo, upper=hi)

def print_stationarity(panel, cols):
    print('\n# Stationarity (ADF & KPSS p-values)')
    for c in cols:
        if c not in panel: continue
        s = panel[c].dropna()
        try: adf_p = adfuller(s)[1]
        except: adf_p = np.nan
        try: kpss_p = kpss(s, nlags='auto')[1]
        except: kpss_p = np.nan
        print(f'{c:<14} ADF={adf_p:.3f}  KPSS={kpss_p:.3f}')

def run_suite(region, txt_path):
    print('\n' + '='*80)
    print(f'>>> {region}  (texts: {txt_path})')
    print('='*80)

    df_m = load_texts_score_sentiment(txt_path)

    cred = load_credibility_planB(CRED_PATH, region, prefer_scaled=True)
    inf_gap, vol_SR = load_macro_series(MACRO_PATH, region)

    panel = (df_m.join(cred.rename('credibility'), how='inner')
                 .join(pd.concat([inf_gap, vol_SR], axis=1), how='inner')
                 .dropna())
    panel_std = pd.DataFrame(StandardScaler().fit_transform(panel),
                             columns=panel.columns, index=panel.index)

    print(f'→ Panel length: {len(panel_std)} months, {panel_std.index.min().date()} → {panel_std.index.max().date()}')

    print_stationarity(panel, ['credibility','finbert_ema6m','hm_ema6m'])

    SENT = 'hm_ema6m'
    y  = panel_std['credibility']
    Xb = panel_std[[SENT,'inf_gap','vol_SR']]

    summary = []

    def add(tag, coef, p):
        summary.append({'region':region, 'test':tag,
                        'beta_or_effect': float(coef) if pd.notna(coef) else np.nan,
                        'p_value': float(p) if pd.notna(p) else np.nan})

    base = nw_ols(y, Xb, lags=nw_lags(len(y)))
    add('Baseline NW-HAC', base.params[SENT], base.pvalues[SENT])
    cl = nw_ols(y, Xb, cov='cluster', groups=y.index.year)
    add('Cluster-robust (year)', cl.params[SENT], cl.pvalues[SENT])

    try:
        rec = sm.OLS(y, sm.add_constant(Xb, has_constant='add')).fit()
        stat, p_cusum = breaks_cusumolsresid(rec.resid, ddof=rec.df_model)
        add('CUSUM stability p', np.nan, p_cusum)
    except Exception:
        add('CUSUM stability p (skipped)', np.nan, np.nan)

    df_w = panel_std.copy()
    for c in ['credibility',SENT,'inf_gap','vol_SR']:
        df_w[c] = winsorize(df_w[c], 0.01)
    res_win = nw_ols(df_w['credibility'], df_w[[SENT,'inf_gap','vol_SR']])
    add('Winsorized 1%', res_win.params[SENT], res_win.pvalues[SENT])

    from statsmodels.tsa.filters.hp_filter import hpfilter
    cyc_y, _  = hpfilter(panel_std['credibility'], lamb=129600)
    cyc_s, _  = hpfilter(panel_std[SENT],       lamb=129600)
    res_hp = nw_ols(cyc_y, pd.DataFrame({SENT:cyc_s,'inf_gap':panel_std['inf_gap'],'vol_SR':panel_std['vol_SR']}))
    add('HP cycles (λ=129600)', res_hp.params[SENT], res_hp.pvalues[SENT])

    D = panel_std[['credibility',SENT,'inf_gap','vol_SR']].diff().rename(
        columns={'credibility':'dcred', SENT:'dsent','inf_gap':'dinf','vol_SR':'dvol'})
    D['ld_dcred'] = D['dcred'].shift(1)
    res_d = nw_ols(D['dcred'], D[['dsent','dinf','dvol','ld_dcred']])
    add('Δ-spec (w/ LD Δy)', res_d.params.get('dsent',np.nan), res_d.pvalues.get('dsent',np.nan))

    r_s = nw_ols(panel_std[SENT], panel_std[['inf_gap','vol_SR']]).resid
    res_r = nw_ols(panel_std['credibility'], pd.DataFrame({f'{SENT}_resid':r_s}))
    add('Residualized sentiment→cred', res_r.params[0], res_r.pvalues[0])

    for s in [SENT.replace('6m','3m'), SENT.replace('6m','12m')]:
        if s in panel_std.columns:
            rr = nw_ols(y, panel_std[[s,'inf_gap','vol_SR']])
            add(f'Alt smoother {s}', rr.params[s], rr.pvalues[s])

    try:
        split = pd.Timestamp('2014-01-01')
        for lab, sl in [('Pre-14', slice(None, split)), ('Post-14', slice(split, None))]:
            y_ = panel_std.loc[sl,'credibility']; X_ = sm.add_constant(panel_std.loc[sl,[SENT,'inf_gap','vol_SR']], has_constant='add')
            m  = sm.OLS(y_, X_, missing='drop').fit(cov_type='HAC', cov_kwds={'maxlags':nw_lags(len(y_))})
            add(f'Subsample {lab}', m.params[SENT], m.pvalues[SENT])
    except Exception:
        pass

    try:
        gt = grangercausalitytests(panel_std[['credibility',SENT]].dropna(), maxlag=12, verbose=False)
        p_min = min(gt[k][0]['ssr_ftest'][1] for k in range(1,13))
        add('Granger (best p in 1..12)', np.nan, p_min)
    except Exception:
        add('Granger (failed)', np.nan, np.nan)

    for L in (3,6):
        ld = panel_std[SENT].shift(-L)
        rL = nw_ols(panel_std['credibility'], pd.DataFrame({'lead':ld,'inf_gap':panel_std['inf_gap'],'vol_SR':panel_std['vol_SR']}))
        add(f'Placebo lead {L}', rL.params.get('lead',np.nan), rL.pvalues.get('lead',np.nan))

    def block_placebo_beta(reps=400, block=12):
        x = panel_std[SENT].values
        yv = y.values
        inf = panel_std['inf_gap'].values
        vol = panel_std['vol_SR'].values
        n = len(x)
        rng = np.random.default_rng(123)
        betas=[]
        for _ in range(reps):
            idx=[]; i=0
            while i<n:
                j = rng.integers(0, max(1, n-block))
                take = min(block, n-i)
                idx.extend(range(j, j+take))
                i += take
            x_star = x[np.array(idx[:n])]
            X = sm.add_constant(np.column_stack([x_star, inf, vol]))
            res = sm.OLS(yv, X, missing='drop').fit(cov_type='HAC', cov_kwds={'maxlags':nw_lags(n)})
            betas.append(res.params[1])
        return np.array(betas)
    betas_null = block_placebo_beta()
    beta_obs   = base.params[SENT]
    p_block    = (np.sum(np.abs(betas_null) >= np.abs(beta_obs)) + 1) / (len(betas_null)+1)
    add('Block-shuffle placebo p', beta_obs, p_block)

    try:
        ar = ARDL(endog=panel_std['credibility'], exog=panel_std[[SENT,'inf_gap','vol_SR']], lags=1, order=1).fit()
        phi = ar.params.get('credibility.L1', np.nan)
        b0  = ar.params.get(SENT, np.nan)
        b1  = ar.params.get(f'{SENT}.L1', np.nan)
        long_run = (b0+b1)/(1-phi) if np.isfinite(phi) and (1-phi)!=0 else np.nan
        add('ARDL long-run approx', long_run, np.nan)
    except Exception:
        add('ARDL long-run approx (skipped)', np.nan, np.nan)

    H = 12; betas_lp=[]; ci_lp=[]
    for k in range(1, H+1):
        yk = panel_std['credibility'].shift(-k)
        Xk = pd.concat([panel_std[[SENT,'inf_gap','vol_SR']],
                        panel_std['credibility'].shift(1).rename('cred_lag')], axis=1)
        d  = pd.concat([yk, Xk], axis=1).dropna()
        res = sm.OLS(d.iloc[:,0], sm.add_constant(d.iloc[:,1:], has_constant='add')).fit(
            cov_type='HAC', cov_kwds={'maxlags':nw_lags(len(d))})
        betas_lp.append(res.params[SENT]); ci_lp.append(1.96*res.bse[SENT])
    betas_lp, ci_lp = np.array(betas_lp), np.array(ci_lp)

    plt.figure(figsize=(6.4,3.0))
    x = np.arange(1,H+1)
    plt.plot(x, betas_lp)
    plt.fill_between(x, betas_lp-ci_lp, betas_lp+ci_lp, alpha=.25)
    plt.axhline(0, ls='--', c='k'); plt.title(f'{region} | LP-IRF  shock({SENT}) → credibility')
    plt.xlabel('months ahead'); plt.tight_layout(); plt.show()

    try:
        var_df = panel_std[[SENT,'credibility']].dropna()
        L = VAR(var_df).select_order(8).aic or 2
        var_res = VAR(var_df).fit(L)
        irf = var_res.irf(12)
        fig = irf.plot(impulse=SENT, response='credibility', orth=False)
        fig.suptitle(f'{region} | VAR-IRF  shock({SENT}) → credibility', y=1.02)
        plt.tight_layout(); plt.show()
    except Exception as e:
        print(f'{region} VAR IRF skipped:', e)

    K = 6; q = 0.90; min_gap = 3
    df_ev = panel_std.copy()
    df_ev['dcred'] = df_ev['credibility'].diff()
    df_ev['dsent'] = df_ev[SENT].diff()
    df_ev['dinf']  = df_ev['inf_gap'].diff()

    thr = df_ev['dinf'].quantile(q)
    cand = df_ev.index[(df_ev['dinf'] >= thr)]
    kept = []
    for t in cand:
        if not kept: kept.append(t)
        elif (t.year - kept[-1].year)*12 + (t.month - kept[-1].month) >= min_gap:
            kept.append(t)
    events = pd.DatetimeIndex(kept)
    print(f"{region} events (q={int(q*100)}% Δinfl):", [d.strftime('%Y-%m') for d in events])

    rows=[]
    for e_id, t0 in enumerate(events):
        for k in range(-K, K+1):
            tk = (t0 + pd.DateOffset(months=k))
            if tk in df_ev.index:
                rows.append({'event': e_id, 'k': k,
                             'dcred': float(df_ev.at[tk,'dcred']),
                             'inf_gap': float(df_ev.at[tk,'inf_gap']),
                             'vol_SR': float(df_ev.at[tk,'vol_SR'])})
    ev = pd.DataFrame(rows).dropna()
    if not ev.empty:
        dK = pd.get_dummies(ev['k'], prefix='k')
        if 'k_-1' in dK.columns: dK = dK.drop(columns=['k_-1'])
        X_ev = pd.concat([dK, ev[['inf_gap','vol_SR']]], axis=1).astype(float)
        mod_ev = sm.OLS(ev['dcred'].astype(float), sm.add_constant(X_ev, has_constant='add')).fit(
            cov_type='cluster', cov_kwds={'groups': ev['event']})
        coef = mod_ev.params.filter(like='k_'); se = mod_ev.bse.filter(like='k_')
        ks = np.array(sorted([int(c.split('_',1)[1]) for c in coef.index]))
        beta = np.array([coef[f'k_{k}'] for k in ks]); bse = np.array([se[f'k_{k}'] for k in ks])

        plt.figure(figsize=(8,3.2))
        plt.plot(ks, beta, marker='o')
        plt.fill_between(ks, beta-1.96*bse, beta+1.96*bse, alpha=.2)
        plt.axhline(0, ls='--', c='k'); plt.axvline(0, ls=':', c='k')
        plt.title(f'{region} | Event-study: ΔCred γύρω από μεγάλες +Δ inflation (baseline k=-1)')
        plt.xlabel('k (months from event)'); plt.tight_layout(); plt.show()

        names = [f'k_{i}' for i in range(0, min(3,K)+1) if f'k_{i}' in coef.index]
        if names:
            V  = mod_ev.cov_params().loc[names, names].values
            b  = float(np.sum([coef[n] for n in names]))
            se = float(np.sqrt(np.ones(len(names)) @ V @ np.ones(len(names))))
            from scipy.stats import norm
            p  = 2*(1-norm.cdf(abs(b/se))) if se>0 else np.nan
            add('Event-study binned (0–3)', b, p)

        pre_cols = [c for c in coef.index if int(c.split('_')[1]) <= -2]
        if pre_cols:
            L = np.zeros((len(pre_cols), len(mod_ev.params)))
            names_all = list(mod_ev.params.index)
            for r, nm in enumerate(pre_cols):
                L[r, names_all.index(nm)] = 1.0
            wld = mod_ev.wald_test(L)
            add('Event-study pre-trend (k≤-2) joint p', np.nan, float(np.atleast_1d(wld.pvalue)[0]))
    else:
        add('Event-study (no events)', np.nan, np.nan)

    out = pd.DataFrame(summary).round(3)
    print('\n' + '-'*60)
    print(f'{region} – ROBUSTNESS SUMMARY')
    print(out[['test','beta_or_effect','p_value']].to_string(index=False))
    print('-'*60 + '\n')
    return out, panel_std

all_summaries = []
panels = {}
for r in ['US','EU','UK']:
    summ, pstd = run_suite(r, TXT_FILES[r])
    all_summaries.append(summ.assign(region=r))
    panels[r] = pstd

full_summary = pd.concat(all_summaries, ignore_index=True)
print('\n====================  FULL ROBUSTNESS SUMMARY (US/EU/UK)  ====================')
print(full_summary[['region','test','beta_or_effect','p_value']].to_string(index=False))
print('===============================================================================')
```
