```python
%pip install -q arch linearmodels ruptures statsmodels
import warnings, logging, pathlib, json, re
import numpy as np, pandas as pd, matplotlib.pyplot as plt
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', message='overflow encountered in exp', category=RuntimeWarning)
warnings.filterwarnings("ignore",
    message="No frequency information was provided, so inferred frequency")
logging.basicConfig(level=logging.INFO, format='[%(levelname)s] %(message)s')
np.random.seed(42)

import statsmodels.api as sm
import statsmodels.tsa.api as smt
from statsmodels.tsa.statespace.structural import UnobservedComponents
from arch import arch_model
from linearmodels.panel import PanelOLS
from scipy import stats as _st
from scipy.linalg import qr as _qr
import ruptures as rpt

CFG = dict(
    FILE   = pathlib.Path('/content/chapter 4 august 2025 1st block.xlsx'),
    ECON   = ['US','EU','UK'],
    START  = '2008-01-01',  END = '2022-01-01',
    AR_MAX = 2,
    MBLOCK = dict(block=12, reps=600),
    H_LIST = [1,3,6,12],
    DK_BW  = 6,
    TARGET = {'US':2.0,'EU':2.0,'UK':2.0},
    ANCHOR = dict(end='2010-01-01', p_lo=5, p_hi=95),
    ECDF_ANCHOR_ENDS = ['2009-12-01','2010-12-01','2011-12-01'],
    FLAGS  = dict(PLOTS=True, FOREC=True, EVENTS=True, ROBUST=True,
                  LP_VALIDATION=True, ECON=True),
    COUNTRY_TRENDS = False,
    BIGMOVE_THRESH = 0.25
)

CFG.update(dict(
    FLAGS_EXTRA=dict(
        TIME_FE_MAIN=True,
        TRENDS_APPENDIX=True,
        EXPECT_GRID=True,
        ALT_CRED_R12_HAR=True,
        BANDS=True,
        MACRO_OOS=True,
        CLASSIFY=True,
        INTERACTIONS=True,
        TARGETED_PLACEBO=True,
        CAMERA_READY=True
    ),
    HALFLIVES=[6,12,24,36],
    BETA_BLEND=0.5,
))

dk_BW = CFG['DK_BW']

def parse_sheet(name):
    key = {'shadow rates':'SR','policy rates':'POLICY','inflation':'INF',
           'volatility':'VOL','composite leader indicator':'CLI'}
    parts = name.split(); econ = parts[-1].upper(); head = ' '.join(parts[:-1]).lower()
    for k,v in key.items():
        if k in head: return v, econ
    if head.startswith('rgdp'): return 'GDP', econ
    return None, None

def to_monthly(s: pd.Series) -> pd.Series:
    s = s.copy()
    if not isinstance(s.index, pd.DatetimeIndex):
        s.index = pd.to_datetime(s.index)
    f = (pd.infer_freq(s.index[:5]) or '').upper()
    if f.startswith(('D','B')):
        s = s.resample('MS').last()
    elif f.startswith('Q'):
        s = s.resample('QS').last().interpolate('linear').resample('MS').ffill()
    else:
        if not s.index.is_month_start.all():
            s.index = s.index.to_period('M').to_timestamp('start')
        s = s.asfreq('MS')
    return s

def ensure_ms_freq(s: pd.Series) -> pd.Series:
    s = pd.Series(s).copy()
    if not isinstance(s.index, pd.DatetimeIndex):
        s.index = pd.to_datetime(s.index)
    return s.asfreq('MS')

def select_ar(y: pd.Series, max_p=2):
    y = ensure_ms_freq(pd.Series(y).dropna())
    best, lag = np.inf, 0
    for p in range(max_p+1):
        try:
            aic = smt.ARIMA(y, order=(p,0,0)).fit(method='innovations_mle', disp=0).aic
            if aic < best: best, lag = aic, p
        except: pass
    return lag

def fit_garch(y: pd.Series, lags_ar: int):
    y = ensure_ms_freq(pd.Series(y).astype(float).dropna())
    if len(y) < 24: return None, y.rolling(12).std()
    mu, sd = y.mean(), y.std(ddof=0)
    if not np.isfinite(sd) or sd == 0: return None, y.rolling(12).std()
    y_std = (y - mu) / sd
    try:
        am = arch_model(y_std, mean='AR', lags=lags_ar, vol='EGARCH', p=1, o=0, q=1, dist='t', rescale=True)
        res = am.fit(disp='off', show_warning=False, options={'maxiter': 2500})
        sig = pd.Series(res.conditional_volatility, index=y.index).replace([np.inf,-np.inf], np.nan)
        if sig.isna().mean() > 0.2:
            logging.warning("EGARCH produced many NaN; fallback rolling 12m std.")
            return None, y.rolling(12).std()
        return res, sig
    except Exception as e:
        logging.warning(f"EGARCH failed: {e}; fallback rolling 12m std.")
        return None, y.rolling(12).std()

def scale_to_0_100(df_rawcred: pd.DataFrame, anchor_spec: dict):
    idx = df_rawcred.index
    base_mask = (idx >= pd.Timestamp(anchor_spec.get('start', idx.min()))) & \
                (idx <= pd.Timestamp(anchor_spec.get('end', idx.max())))
    base = df_rawcred.loc[base_mask].stack()
    p_lo = np.nanpercentile(base.values, anchor_spec.get('p_lo',5))
    p_hi = np.nanpercentile(base.values, anchor_spec.get('p_hi',95))
    width = max(1e-8, p_hi - p_lo)
    scaled = (df_rawcred - p_lo) / width * 100.0
    return scaled.clip(lower=0, upper=100), dict(p_lo=p_lo, p_hi=p_hi)

def prob_cred_from_ecdf(vol_sr_df: pd.DataFrame, anchor_end: str, pooled=True) -> pd.DataFrame:
    anchor_end = pd.Timestamp(anchor_end)
    V = vol_sr_df.copy()
    V.columns = [str(c).split('_')[0] for c in V.columns]
    out = {}
    if pooled:
        base_vals = []
        for e in V.columns:
            s = V[e].dropna()
            base_vals.append((-s.loc[s.index <= anchor_end]).values)
        base = np.concatenate(base_vals) if base_vals else np.array([])
        if base.size < 12:
            base = np.sort(np.concatenate([(-V[e].dropna().values) for e in V.columns]))[:max(12, base.size)]
        def ecdf(x):
            if base.size == 0: return np.nan
            return (base <= x).mean()
        for e in V.columns:
            s = V[e].dropna()
            out[e] = s.apply(lambda v: ecdf(-v) * 100.0)
    else:
        for e in V.columns:
            s = V[e].dropna()
            base = (-s.loc[s.index <= anchor_end]).values
            if base.size < 12:
                base = (-s.iloc[:max(12, len(s)//5)].values)
            base = np.asarray(base)
            def ecdf(x): return (base <= x).mean() if base.size>0 else np.nan
            out[e] = s.apply(lambda v: ecdf(-v) * 100.0)
    prob_df = pd.DataFrame(out)
    prob_df.columns = [f'prob_{c}' for c in prob_df.columns]
    return prob_df

def build_expectations_proxy(inf: pd.Series, halflife=12):
    s = ensure_ms_freq(pd.Series(inf).astype(float))
    ew = s.ewm(halflife=halflife, adjust=False, min_periods=max(3, halflife//3)).mean()
    try:
        uc = UnobservedComponents(s, level='llevel')
        res = uc.fit(disp=False)
        try:
            lvl = pd.Series(res.level.smoothed, index=s.index)
        except Exception:
            lvl = pd.Series(res.filtered_state[0], index=s.index)
        proxy = 0.5*ew + 0.5*lvl
        proxy.name = 'exp_INF'
        return proxy
    except Exception:
        return ew.rename('exp_INF')

def realized_vol(s: pd.Series, H: int, variance=False):
    s = ensure_ms_freq(pd.Series(s))
    rstd = s.shift(-1).rolling(H, min_periods=H).std(ddof=0)
    return (rstd**2) if variance else rstd

def qlike_loss(y_var, f_var, eps=1e-10):
    y = pd.Series(y_var).astype(float).clip(lower=eps)
    f = pd.Series(f_var).astype(float).clip(lower=eps)
    return np.log(f) + (y / f)

def har_mix(s: pd.Series):
    s = ensure_ms_freq(pd.Series(s))
    s_lag = s.shift(1)
    v3  = s_lag.rolling(3,  min_periods=3 ).std(ddof=0)
    v6  = s_lag.rolling(6,  min_periods=6 ).std(ddof=0)
    v12 = s_lag.rolling(12, min_periods=12).std(ddof=0)
    return pd.concat([v3, v6, v12], axis=1).mean(axis=1)

def dm_test_hln(loss_diff: pd.Series, h=1, lags=None):
    d = pd.Series(loss_diff).dropna()
    T = len(d)
    if T < 10:
        return np.nan, np.nan, np.nan, np.nan, T
    L = max(0, (h-1) if lags is None else int(lags))
    mu = d.mean()
    def acov(k):
        x = d.iloc[k:] - mu
        y = d.shift(k).dropna() - mu
        n = min(len(x), len(y))
        return 0.0 if n <= 1 else float(np.sum(x.iloc[:n]*y.iloc[:n]) / n)
    var = acov(0)
    for k in range(1, L+1):
        var += 2*(1 - k/(L+1))*acov(k)
    var = var / T
    if var <= 0 or not np.isfinite(var):
        return np.nan, np.nan, np.nan, np.nan, T
    t_dm = mu / np.sqrt(var)
    p_dm = 2*(1 - _st.norm.cdf(abs(t_dm)))
    c = ((T + 1 - 2*h + h*(h-1)/T) / T) if T > 0 else 1.0
    t_hln = t_dm / np.sqrt(c) if c > 0 else np.nan
    p_hln = 2*(1 - _st.t.cdf(abs(t_hln), df=max(1, T-1))) if np.isfinite(t_hln) else np.nan
    return float(t_dm), float(p_dm), float(t_hln), float(p_hln), T

def gw_test_conditional(loss_diff: pd.Series, Z: pd.DataFrame, lags=None):
    d = pd.Series(loss_diff).astype(float).rename('d')
    Z = pd.DataFrame(Z).astype(float)
    df = pd.concat([d, Z], axis=1).replace([np.inf, -np.inf], np.nan).dropna()
    T = len(df)
    if T < 10:
        return np.nan, np.nan, np.nan, np.nan, T
    X = df.drop(columns=['d'])
    keep = X.columns[X.nunique(dropna=True) > 1]
    X = X[keep]
    X = sm.add_constant(X, has_constant='add')
    L = int(lags) if lags is not None else max(1, int(T**(1/3)))
    res = sm.OLS(df['d'].values, X.values).fit(cov_type='HAC', cov_kwds={'maxlags': L})
    q = X.shape[1]
    if q > 1:
        F = res.f_test(np.eye(q))
        F_cpa, p_cpa = float(F.fvalue), float(F.pvalue)
    else:
        F_cpa, p_cpa = np.nan, np.nan
    t_alpha, p_alpha = float(res.tvalues[0]), float(res.pvalues[0])
    return F_cpa, p_cpa, t_alpha, p_alpha, T

def build_expectations_grid(inf_col: pd.Series, halflives=(6,12,24,36)):
    """Επιστρέφει dict με proxies: {name: Series} – EWMA_h, UC (llevel), BLEND_h."""
    s = ensure_ms_freq(pd.Series(inf_col).astype(float))
    out = {}
    try:
        uc = UnobservedComponents(s, level='llevel').fit(disp=False)
        try:
            lvl = pd.Series(uc.level.smoothed, index=s.index)
        except Exception:
            lvl = pd.Series(uc.filtered_state[0], index=s.index)
        out['UC'] = lvl.rename('exp_INF_UC')
    except Exception:
        out['UC'] = s.ewm(halflife=12, adjust=False).mean().rename('exp_INF_UC')
    for h in halflives:
        ew = s.ewm(halflife=h, adjust=False, min_periods=max(3,h//3)).mean().rename(f'exp_INF_EWMA{h}')
        bl = (CFG.get('BETA_BLEND',0.5))*out['UC'] + (1-CFG.get('BETA_BLEND',0.5))*ew
        bl = bl.rename(f'exp_INF_BLEND{h}')
        out[ew.name] = ew; out[bl.name] = bl
    return out

def alt_cred_from_r12_har(sr_series: pd.Series, anchor_spec: dict):
    """Φτιάχνει -logσ̂ από R12/HAR και τα κλιμακώνει 0–100 με ίδιο anchor."""
    s = ensure_ms_freq(sr_series)
    r12 = s.shift(1).rolling(12, min_periods=12).std(ddof=0).rename('R12')
    har = har_mix(s).rename('HAR')
    sig = pd.concat([r12, har], axis=1)
    raw = -np.log(sig.clip(lower=1e-10))
    scaled, statsA = scale_to_0_100(raw, anchor_spec)
    scaled.columns = [f'cred_sc_{c}_ALT' for c in scaled.columns]
    return raw, scaled

def stationary_block_indices(T, block, rng):
    """Επιστρέφει index μήκους T με stationary bootstrap (Politis–Romano “geom blocks”)."""
    p = 1.0/block
    idx = np.empty(T, dtype=int)
    t=0
    while t<T:
        start = rng.integers(0, T)
        L=1
        while t<T and rng.random()>p:
            L+=1
        L = min(L, T-t)
        idx[t:t+L] = (start + np.arange(L)) % T
        t += L
    return idx

def cred_bands_mbb(cred_series: pd.Series, block=12, reps=600, q=(5,95), seed=123):
    """Pointwise bands για μία χώρα (σε ήδη κλιμακωμένο 0–100 cred)."""
    v = cred_series.dropna().values
    T = len(v)
    rng = np.random.default_rng(seed)
    B = np.empty((reps, T))
    for b in range(reps):
        idx = stationary_block_indices(T, block, rng)
        B[b,:] = v[idx]
    low, hi = np.percentile(B, q, axis=0)
    out = pd.DataFrame({'low':low, 'high':hi}, index=cred_series.dropna().index)
    return out

def auc_from_scores(y_true, y_score):
    """AUC χωρίς sklearn – ισοδύναμο του Mann–Whitney U / ROC AUC."""
    y = np.asarray(y_true).astype(int)
    s = np.asarray(y_score).astype(float)
    pos = s[y==1]; neg = s[y==0]
    if len(pos)==0 or len(neg)==0: return np.nan
    r = pd.Series(np.r_[pos, neg]).rank(method='average').values
    n1, n0 = len(pos), len(neg)
    sum_ranks_pos = r[:n1].sum()
    auc = (sum_ranks_pos - n1*(n1+1)/2) / (n1*n0)
    return float(auc)

def calibration_table(y_true, p_hat, bins=10):
    df = pd.DataFrame({'y':y_true, 'p':p_hat}).dropna()
    df['bin'] = pd.qcut(df['p'], q=bins, duplicates='drop')
    g = df.groupby('bin')
    return pd.DataFrame({'pred':g['p'].mean(), 'obs':g['y'].mean(), 'n':g.size()})

def run_time_fe_baseline(df_panel, y='abs_y', xcols=None, dk_bw=6):
    d = df_panel.copy()
    if xcols is None: xcols = baseline_X(d, CFG)
    data = d[[y]+[c for c in xcols if c in d.columns]].dropna().reset_index().set_index(['econ','date']).sort_index()
    yv = data[y]
    X  = data[[c for c in xcols if c in data.columns]]
    X = X.drop(columns=[c for c in X.columns if c.startswith('m_')], errors='ignore')
    X = X.loc[:, ~X.columns.duplicated(keep='first')]
    X = X.drop(columns=[c for c in X.columns if X[c].nunique()<=1], errors='ignore')
    X, _ = drop_collinear_qr(X)
    yv = yv.loc[X.index]
    mod = PanelOLS(yv, X, entity_effects=True, time_effects=True)
    return mod.fit(cov_type='kernel', kernel='bartlett', bandwidth=dk_BW)

def oos_gap_forecasts(df_e: pd.DataFrame, H_list=(1,), start_oos=None):
    """
    Expanding-window προβλέψεις για abs_y. Benchmarks: RW/persistence, AR(1).
    Μοντέλο: OLS(abs_y_{t+H}) ~ cred_t + CLI_t + dSR_t. Επιστρέφει losses & DM/CPA.
    """
    out_rows = []
    s = df_e.sort_index()
    if start_oos is None:
        start_oos = s.index[max(24, int(len(s)*0.5))]
    for H in H_list:
        y = s['abs_y'].shift(-H)
        cred = s['cred_sc_OWN']; cli = s['CLI']; dsr = s['dSR']
        rw_pred = s['abs_y']
        ar_pred = pd.Series(index=s.index, dtype=float)
        for t in s.index[s.index.get_loc(start_oos):]:
            tr = s.loc[:t - pd.offsets.MonthBegin(1)]
            if tr['abs_y'].dropna().shape[0] < 30: continue
            yy = tr['abs_y'].shift(-H)
            X = sm.add_constant(tr['abs_y'])
            try:
                res = sm.OLS(yy.dropna(), X.loc[yy.dropna().index]).fit()
                ar_pred.loc[t] = float(res.predict([1.0, tr['abs_y'].iloc[-1]]))
            except Exception:
                pass
        cred_pred = pd.Series(index=s.index, dtype=float)
        for t in s.index[s.index.get_loc(start_oos):]:
            tr = s.loc[:t - pd.offsets.MonthBegin(1)]
            df = pd.DataFrame({'y':tr['abs_y'].shift(-H),'cred':tr['cred_sc_OWN'],'CLI':tr['CLI'],'dSR':tr['dSR']}).dropna()
            if df.shape[0] < 40: continue
            X = sm.add_constant(df[['cred','CLI','dSR']])
            try:
                res = sm.OLS(df['y'], X).fit()
                cred_pred.loc[t] = float(res.predict([1.0, cred.loc[t], cli.loc[t], dsr.loc[t]]))
            except Exception:
                pass
        target = y.loc[cred_pred.dropna().index]
        L_rw   = (target - rw_pred.loc[target.index])**2
        L_ar   = (target - ar_pred.loc[target.index])**2
        L_cr   = (target - cred_pred.loc[target.index])**2
        for name, Lb in [('RW',L_rw), ('AR1',L_ar)]:
            d_mse = (L_cr - Lb).dropna()
            t_dm, p_dm, t_hln, p_hln, T = dm_test_hln(d_mse, h=H, lags=H-1)
            Z = make_instruments(s, 'SR', realized_vol(s['SR'], H, variance=False))
            F_cpa, p_cpa, t_u, p_u, _ = gw_test_conditional(d_mse, Z, lags=H-1)
            out_rows.append([H, name, T, d_mse.mean(), t_dm, p_dm, t_hln, p_hln, F_cpa, p_cpa, t_u, p_u])
    cols = ['H','vs','T','avg_dMSE','DM_t','DM_p','HLN_t','HLN_p','GW_F','GW_p','GW_tα','GW_pα']
    return pd.DataFrame(out_rows, columns=cols)

def country_swap_placebo_broadcast(panel: pd.DataFrame, K=200, dk_bw=6):
    """
    Placebo στο US broadcast: ανακατεύουμε στον χρόνο το cred_sc_US_bcast
    και κρατάμε τους συντελεστές US_for_EU & US_for_UK.
    """
    if 'cred_sc_US_bcast' not in panel.columns:
        raise KeyError("cred_sc_US_bcast not found.")
    us = panel.xs('US', level='econ')['cred_sc_US_bcast'].dropna().sort_index()
    dates = us.index
    rows=[]
    for _ in range(K):
        sh = us.sample(frac=1.0, replace=False).values
        us_sh = pd.Series(sh, index=dates, name='cred_sc_US_bcast_shuf')
        tmp = (panel.reset_index().merge(us_sh.reset_index(), on='date', how='left')
               .set_index(['date','econ']).sort_index())
        tmp['US_for_EU']=0.0; tmp['US_for_UK']=0.0
        tmp.loc[(slice(None),'EU'),'US_for_EU']=tmp.loc[(slice(None),'EU'),'cred_sc_US_bcast_shuf']
        tmp.loc[(slice(None),'UK'),'US_for_UK']=tmp.loc[(slice(None),'UK'),'cred_sc_US_bcast_shuf']
        res = run_fe(tmp, 'abs_y', baseline_X(tmp, CFG), dk_bw=dk_BW, time_fe=False)
        rows.append({'b_USforEU': float(res.params.get('US_for_EU', np.nan)),
                     'b_USforUK': float(res.params.get('US_for_UK', np.nan))})
    return pd.DataFrame(rows)

def _winner(cmp_label: str, tstat):
    """
    cmp_label: 'ModelA vs ModelB'
    DM t < 0 => A wins; DM t > 0 => B wins.
    """
    try:
        a, b = [x.strip() for x in cmp_label.split('vs')]
    except Exception:
        a, b = "A", "B"
    return a if (pd.notna(tstat) and np.isfinite(tstat) and float(tstat) < 0) else b

def baseline_X(df, cfg):
    cols = ['cred_sc_OWN','US_for_EU','US_for_UK','CLI','dSR']
    cols += [c for c in df.columns if c.startswith('m_')]
    if cfg['COUNTRY_TRENDS']:
        cols += [c for c in df.columns if c.startswith('trend_')]
    return cols

def drop_collinear_qr(X: pd.DataFrame, tol=1e-10):
    X = pd.DataFrame(X).copy()
    X = X.loc[:, ~X.columns.duplicated(keep='first')]
    const_like = [c for c in X.columns if X[c].nunique(dropna=True) <= 1]
    X = X.drop(columns=const_like, errors='ignore')
    if X.shape[1] == 0:
        return X, []
    arr = np.asarray(X.replace([np.inf,-np.inf], np.nan).dropna(how='any'), dtype=float)
    X = X.loc[X.dropna(how='any').index]
    Q, R, piv = _qr(arr, mode='economic', pivoting=True)
    diag = np.abs(np.diag(R)) if R.size else np.array([])
    rank = int((diag > tol).sum()) if diag.size else 0
    keep_idx = list(piv[:rank])
    keep_cols = list(X.columns[keep_idx])
    dropped = [c for c in X.columns if c not in keep_cols]
    return X[keep_cols], dropped

def run_fe(df, ycol, xcols, dk_bw=6, time_fe=False):
    data = df[[ycol] + [c for c in xcols if c in df.columns]].dropna().reset_index().set_index(['econ','date']).sort_index()
    if data.empty:
        raise ValueError("FE input after dropna is empty – check regressors/NaNs.")
    y = data[ycol]
    X = data[[c for c in xcols if c in data.columns]]

    if time_fe:
        X = X.drop(columns=[c for c in X.columns if c.startswith('m_')], errors='ignore')

    X = X.loc[:, ~X.columns.duplicated(keep='first')]
    X = X.drop(columns=[c for c in X.columns if X[c].nunique()<=1], errors='ignore')
    Xq, _ = drop_collinear_qr(X)
    if Xq.shape[1] == 0:
        raise ValueError("All regressors were dropped after QR collinearity guard.")
    y = y.loc[Xq.index]

    mod = PanelOLS(y, Xq, entity_effects=True, time_effects=time_fe)
    return mod.fit(cov_type='kernel', kernel='bartlett', bandwidth=dk_BW)

def make_instruments(df_e: pd.DataFrame, var: str, rv_std: pd.Series) -> pd.DataFrame:
    Z = pd.DataFrame(index=df_e.index)
    Z['dCLI'] = df_e['CLI'].diff()
    if var == 'SR':
        Z['abs_dSR']  = df_e['SR'].diff().abs()
    else:
        Z['abs_dINF'] = df_e['INF'].diff().abs()
    Z['rv_lag1'] = rv_std.shift(1)
    return Z

def fit_gjr_garch(y: pd.Series, lags_ar: int):
    y = ensure_ms_freq(pd.Series(y).astype(float).dropna())
    if len(y) < 24: return None, y.rolling(12).std()
    mu, sd = y.mean(), y.std(ddof=0)
    if not np.isfinite(sd) or sd == 0: return None, y.rolling(12).std()
    y_std = (y - mu) / sd
    try:
        am = arch_model(y_std, mean='AR', lags=lags_ar, vol='GARCH', p=1, o=1, q=1, dist='t', rescale=True)
        res = am.fit(disp='off', show_warning=False, options={'maxiter': 2500})
        sig = pd.Series(res.conditional_volatility, index=y.index).replace([np.inf,-np.inf], np.nan)
        if sig.isna().mean() > 0.2:
            logging.warning("GJR-GARCH produced many NaN; fallback rolling 12m std.")
            return None, y.rolling(12).std()
        return res, sig
    except Exception as e:
        logging.warning(f"GJR-GARCH failed: {e}; fallback rolling 12m std.")
        return None, y.rolling(12).std()

def pca_decommonize(df_sigma: pd.DataFrame):
    X = df_sigma.copy()
    X = X.dropna(how='all')
    if X.shape[1] < 2 or X.dropna().empty:
        return df_sigma
    Xc = (X - X.mean())/X.std(ddof=0)
    Xc = Xc.fillna(method='ffill').fillna(method='bfill')
    U, S, Vt = np.linalg.svd(Xc.values, full_matrices=False)
    pc1 = np.outer(U[:,0]*S[0], Vt[0,:])
    resid = Xc - pc1
    out = resid * X.std(ddof=0) + X.mean()
    return out.reindex(df_sigma.index)

def build_scaled_from_sigma(sig_df: pd.DataFrame, anchor_spec: dict, prefix: str):
    cred_raw_loc = (-np.log(sig_df)).copy()
    scaled, statsA = scale_to_0_100(cred_raw_loc, anchor_spec)
    scaled.columns = [f'cred_sc_{c}_{prefix}' for c in scaled.columns]
    return cred_raw_loc, scaled, statsA

def expanding_oos_sigma(series: pd.Series, model='EGARCH', start_oos=None, ar_max=2, horizon=1):
    s = ensure_ms_freq(pd.Series(series).dropna().sort_index())
    if start_oos is None:
        start_oos = s.index[max(24, int(len(s)*0.5))]
    out = pd.Series(index=s.index, dtype=float)
    for t in s.index[s.index.get_loc(start_oos):]:
        s_train = s.loc[:t - pd.offsets.MonthBegin(1)]
        if len(s_train) < 24: continue
        p = select_ar(s_train, ar_max)
        mu, sd = s_train.mean(), s_train.std(ddof=0)
        if not np.isfinite(sd) or sd == 0:
            out.loc[t] = np.nan; continue
        y_std = (s_train - mu)/sd
        try:
            if model=='EGARCH':
                am = arch_model(y_std, mean='AR', lags=p, vol='EGARCH', p=1, o=0, q=1, dist='t', rescale=True)
            elif model=='GJR':
                am = arch_model(y_std, mean='AR', lags=p, vol='GARCH', p=1, o=1, q=1, dist='t', rescale=True)
            else:
                raise ValueError("model must be 'EGARCH' or 'GJR'")
            res = am.fit(disp='off', show_warning=False)
            f = res.forecast(horizon=horizon).variance.iloc[-1, -1]
            out.loc[t] = float(np.sqrt(max(f, 1e-12))) * sd
        except Exception:
            out.loc[t] = np.nan
    return out

def mcs_hln(loss_df: pd.DataFrame, alpha=0.10, B=400, block=6, seed=123):
    rng = np.random.default_rng(seed)
    S = list(loss_df.columns)
    T = loss_df.shape[0]
    def boot_means(L):
        idx = np.arange(T)
        means = []
        for _ in range(B):
            pos = rng.integers(0, T, size=int(np.ceil(T/block)))
            bidx = np.hstack([ (idx[p: p+block] if p+block<=T else np.r_[idx[p:], idx[:(p+block)%T]]) for p in pos ])[:T]
            means.append(L[bidx,:].mean(axis=0))
        return np.vstack(means)
    while len(S) > 1:
        L = loss_df[S].to_numpy()
        m_hat = L.mean(axis=0)
        bmeans = boot_means(L)
        worst = S[int(np.argmax(m_hat))]
        d_worst = bmeans[:, np.argmax(m_hat)] - bmeans.min(axis=1)
        p_worst = np.mean(d_worst > d_worst.mean())
        if p_worst < alpha:
            S.remove(worst)
        else:
            break
    ranks = loss_df[S].mean().sort_values().rank().to_dict()
    return S, ranks

def build_dynamic_ecdf_prob(vol_df: pd.DataFrame, ends: list):
    out = {}
    for a in ends:
        pr = prob_cred_from_ecdf(vol_df, anchor_end=a, pooled=True)
        for c in pr.columns:
            out[(a, c.replace('prob_',''))] = pr[c]
    return pd.concat(out, axis=1)

def country_swap_placebo(panel: pd.DataFrame, K=100, dk_bw=6):
    if 'cred_sc_US_bcast' not in panel.columns:
        raise KeyError("cred_sc_US_bcast not found in panel (broadcast US credibility).")
    us = panel.xs('US', level='econ')['cred_sc_US_bcast'].dropna().sort_index()
    dates = us.index
    rows = []
    for _ in range(K):
        shuf = us.sample(frac=1.0, replace=False).values
        us_sh = pd.Series(shuf, index=dates, name='cred_sc_US_bcast_shuf')
        tmp = (panel.reset_index().merge(us_sh.reset_index(), on='date', how='left')
               .set_index(['date','econ']).sort_index())
        tmp['US_for_EU'] = 0.0; tmp['US_for_UK'] = 0.0
        tmp.loc[(slice(None),'EU'),'US_for_EU'] = tmp.loc[(slice(None),'EU'),'cred_sc_US_bcast_shuf']
        tmp.loc[(slice(None),'UK'),'US_for_UK'] = tmp.loc[(slice(None),'UK'),'cred_sc_US_bcast_shuf']
        res = run_fe(tmp, 'abs_y', baseline_X(tmp, CFG), dk_bw=dk_BW, time_fe=False)
        rows.append({
            'b_USforEU': float(res.params.get('US_for_EU', np.nan)),
            'b_USforUK': float(res.params.get('US_for_UK', np.nan))
        })
    return pd.DataFrame(rows)

def run_iv_panel_manual(panel: pd.DataFrame, dk_bw=6):
    """
    Manual 2SLS για panel με entity FE (Driscoll–Kraay).
    Ενδογενές: cred_sc_OWN
    Όργανα: IV1 = BigMove_t × sign(new_y_{t-1}), BigMove_t
    """
    df = panel.reset_index().sort_values(['econ','date']).copy()
    df['BigMove'] = (df.groupby('econ')['POLICY'].diff().abs() > CFG['BIGMOVE_THRESH']).astype(int)
    df['sign_gap_lag1'] = np.sign(df.groupby('econ')['new_y'].shift(1))
    df['IV1'] = df['BigMove'] * df['sign_gap_lag1']
    mcols = [c for c in df.columns if c.startswith('m_')]
    keep = ['abs_y','cred_sc_OWN','US_for_EU','US_for_UK','CLI','dSR','IV1','BigMove'] + mcols + ['econ','date']
    d = (df[keep].dropna().set_index(['econ','date']).sort_index())

    X1 = pd.concat([d[['IV1','BigMove','US_for_EU','US_for_UK','CLI','dSR']], d[mcols]], axis=1)
    X1, _ = drop_collinear_qr(X1)
    y1 = d['cred_sc_OWN'].loc[X1.index]
    mod1 = PanelOLS(y1, X1, entity_effects=True)
    res1 = mod1.fit(cov_type='kernel', kernel='bartlett', bandwidth=dk_BW)
    d = d.assign(cred_hat=res1.fitted_values)

    X2 = pd.concat([d[['cred_hat','US_for_EU','US_for_UK','CLI','dSR']], d[mcols]], axis=1)
    X2, _ = drop_collinear_qr(X2)
    y2 = d['abs_y'].loc[X2.index]
    mod2 = PanelOLS(y2, X2, entity_effects=True)
    res2 = mod2.fit(cov_type='kernel', kernel='bartlett', bandwidth=dk_BW)
    return res1, res2

def fair_horserace_and_mcs(panel: pd.DataFrame, vol_sr_df: pd.DataFrame, H_LIST, start_oos='2012-01-01'):
    results = []
    mcs_rows = []
    for econ in CFG['ECON']:
        df_e = panel.reset_index().set_index('date').query("econ==@econ").copy().sort_index()

        sig_eg_oos = expanding_oos_sigma(df_e['SR'], model='EGARCH', start_oos=pd.Timestamp(start_oos))
        sig_gjr_oos= expanding_oos_sigma(df_e['SR'], model='GJR',    start_oos=pd.Timestamp(start_oos))
        sig_r12    = df_e['SR'].shift(1).rolling(12, min_periods=12).std(ddof=0)
        sig_har    = har_mix(df_e['SR'])
        for H in H_LIST:
            rv_std = realized_vol(df_e['SR'], H, variance=False)
            rv_var = realized_vol(df_e['SR'], H, variance=True)
            models = {'EGARCH_OOS': sig_eg_oos, 'GJR_OOS': sig_gjr_oos, 'R12': sig_r12, 'HAR': sig_har}
            L_mse, L_ql = {}, {}
            for m, fstd in models.items():
                fstd = fstd.reindex(df_e.index)
                L_mse[m] = (rv_std - fstd)**2
                fvar = (fstd**2).clip(lower=1e-10)
                L_ql[m]  = qlike_loss(rv_var, fvar)
            Lm = pd.DataFrame(L_mse).dropna(how='all').dropna(axis=0)
            Lq = pd.DataFrame(L_ql ).dropna(how='all').dropna(axis=0)
            base = 'R12'
            for m in models:
                if m==base: continue
                d_mse = Lm[m]-Lm[base]
                d_ql  = Lq[m]-Lq[base]
                t_dm, p_dm, t_hln, p_hln, T = dm_test_hln(d_mse, h=H, lags=H-1)
                Z = make_instruments(df_e, 'SR', rv_std)
                F_cpa, p_cpa, t_u, p_u, _ = gw_test_conditional(d_mse, Z, lags=H-1)
                cmp = f'{m} vs {base}'
                results.append([econ,'SR',H,'MSE',cmp,T,t_dm,p_dm,t_hln,p_hln,_winner(cmp,t_dm),F_cpa,p_cpa,t_u,p_u])
                t_dm, p_dm, t_hln, p_hln, T = dm_test_hln(d_ql, h=H, lags=H-1)
                F_cpa, p_cpa, t_u, p_u, _ = gw_test_conditional(d_ql, Z, lags=H-1)
                results.append([econ,'SR',H,'QLIKE',cmp,T,t_dm,p_dm,t_hln,p_hln,_winner(cmp,t_dm),F_cpa,p_cpa,t_u,p_u])
            if Lm.shape[1] >= 2:
                survivors_mse, _ = mcs_hln(Lm.dropna(), alpha=0.10, B=400, block=max(2,H))
                mcs_rows.append([econ,'SR',H,'MSE',';'.join(survivors_mse), Lm.mean().to_dict()])
            if Lq.shape[1] >= 2:
                survivors_ql , _ = mcs_hln(Lq.dropna(), alpha=0.10, B=400, block=max(2,H))
                mcs_rows.append([econ,'SR',H,'QLIKE',';'.join(survivors_ql ), Lq.mean().to_dict()])
    cols = ['econ','var','H','loss','cmp','T','DM_t','DM_p','HLN_t','HLN_p','Winner_DM','GW_F','GW_p','GW_tα','GW_pα']
    dm_oos = pd.DataFrame(results, columns=cols)
    mcs_tab = pd.DataFrame(mcs_rows, columns=['econ','var','H','loss','survivors','avg_loss'])
    return dm_oos, mcs_tab

raw = {}
xls = pd.read_excel(CFG['FILE'], sheet_name=None)
for sheet, df in xls.items():
    var, econ = parse_sheet(sheet)
    if var and econ in CFG['ECON']:
        df = df.iloc[:, :2].copy(); df.columns = ['date', var]; df['date'] = pd.to_datetime(df['date'])
        s = to_monthly(df.set_index('date')[var]); raw[(econ,var)] = s.loc[CFG['START']:CFG['END']]
panel = (pd.concat(raw, axis=1)
           .swaplevel(0,1, axis=1).sort_index(axis=1)
           .stack(level=1).rename_axis(['date','econ']).sort_index())

print("Fitting EGARCH on SR:")
vol_cols_sr = {}
for econ, g in panel['SR'].groupby('econ'):
    y = g.droplevel('econ').dropna(); arlag = select_ar(y, CFG['AR_MAX'])
    _, sig = fit_garch(y, arlag)
    vol_cols_sr[f'{econ}_EGARCH_t'] = sig
vol_sr_df = pd.concat(vol_cols_sr, axis=1)
vol_sr_ser = vol_sr_df.stack().rename_axis(['date','spec']).sort_index()

print("Fitting EGARCH on INF:")
vol_cols_inf = {}
for econ, g in panel['INF'].groupby('econ'):
    y = g.droplevel('econ').dropna(); arlag = select_ar(y, CFG['AR_MAX'])
    _, sig = fit_garch(y, arlag)
    vol_cols_inf[f'{econ}_EGARCH_t'] = sig
vol_inf_df = pd.concat(vol_cols_inf, axis=1)

cred_raw = (-np.log(vol_sr_ser)).unstack('spec')
scaled_all, anchor_stats = scale_to_0_100(cred_raw, CFG['ANCHOR'])
scaled_all.columns = ['cred_sc_'+c for c in scaled_all.columns]

def build_prob_panel(anchor_end: str, base_panel: pd.DataFrame) -> pd.DataFrame:
    prob_df = prob_cred_from_ecdf(vol_sr_df, anchor_end=anchor_end, pooled=True)
    tmp = (base_panel.reset_index()
           .merge(prob_df.reset_index(), on='date', how='left'))
    tmp['cred_sc_OWN'] = np.select(
        [tmp['econ'].eq('US'), tmp['econ'].eq('EU'), tmp['econ'].eq('UK')],
        [tmp['prob_US'],        tmp['prob_EU'],        tmp['prob_UK']],
        default=np.nan
    )
    tmp['cred_sc_US_bcast'] = tmp['prob_US']
    tmp['US_for_EU'] = np.where(tmp['econ'].eq('EU'), tmp['cred_sc_US_bcast'], 0.0)
    tmp['US_for_UK'] = np.where(tmp['econ'].eq('UK'), tmp['cred_sc_US_bcast'], 0.0)
    return tmp.set_index(['date','econ']).sort_index()

panel = (panel.reset_index()
              .merge(scaled_all.reset_index(), on='date', how='left')
              .set_index(['date','econ']).sort_index())

if CFG['FLAGS_EXTRA']['ALT_CRED_R12_HAR']:
    alt_list = []
    for econ, g in panel['SR'].groupby('econ'):
        _, sc_alt = alt_cred_from_r12_har(g.droplevel('econ'), CFG['ANCHOR'])
        base = [c.replace('cred_sc_', '') for c in sc_alt.columns]
        sc_alt.columns = [f'cred_sc_{econ}_{b}' for b in base]
        alt_list.append(sc_alt)

    cred_alt_df = pd.concat(alt_list, axis=1).sort_index()
    cred_alt_df.index.name = 'date'

    panel = (panel.reset_index()
                  .merge(cred_alt_df.reset_index(), on='date', how='left')
                  .set_index(['date','econ']).sort_index())

inf_wide = panel['INF'].unstack('econ')
exp_wide = inf_wide.apply(build_expectations_proxy, axis=0)
panel = panel.join(exp_wide.stack().rename('exp_INF'))
_targets = pd.Series(CFG['TARGET'])
panel['target'] = panel.index.get_level_values('econ').map(_targets)
panel['new_y']  = panel['exp_INF'] - panel['target']
panel['abs_y']  = panel['new_y'].abs()

exp_grid_tabs = {}
if CFG['FLAGS_EXTRA']['EXPECT_GRID']:
    print("\n=== Expectations grid (EWMA/UC/Blend) ===")
    for econ in CFG['ECON']:
        g = panel.xs(econ, level='econ')['INF']
        proxies = build_expectations_grid(g, CFG['HALFLIVES'])
        exp_grid_tabs[econ] = pd.concat(proxies, axis=1)
    alt = panel.copy().reset_index()
    alt = alt.merge(exp_grid_tabs['US'].reset_index().rename(columns={'index':'date'}), on='date', how='left')

panel['dSR'] = panel.groupby('econ')['SR'].diff()

us_sc = (panel.xs('US', level='econ')['cred_sc_US_EGARCH_t']
         .rename('cred_sc_US_bcast').to_frame().reset_index())
panel = (panel.reset_index()
              .merge(us_sc, on='date', how='left')
              .set_index(['date','econ']).sort_index())

panel['cred_sc_OWN'] = np.nan
for e in CFG['ECON']:
    col = f'cred_sc_{e}_EGARCH_t'
    if col in panel.columns:
        panel.loc[(slice(None), e), 'cred_sc_OWN'] = panel.loc[(slice(None), e), col]
panel['US_for_EU'] = 0.0; panel['US_for_UK'] = 0.0
panel.loc[(slice(None),'EU'), 'US_for_EU'] = panel.loc[(slice(None),'EU'), 'cred_sc_US_bcast']
panel.loc[(slice(None),'UK'), 'US_for_UK'] = panel.loc[(slice(None),'UK'), 'cred_sc_US_bcast']

panel = panel.reset_index()
panel['m'] = panel['date'].dt.month
mD = pd.get_dummies(panel['m'], prefix='m', dtype=int)
if 'm_1' in mD.columns: mD.drop(columns=['m_1'], inplace=True)
panel = pd.concat([panel, mD], axis=1)
if CFG['COUNTRY_TRENDS']:
    for econ in CFG['ECON']:
        mask = panel['econ']==econ
        panel.loc[mask, f'trend_{econ}'] = np.arange(mask.sum(), dtype=float)
panel = panel.set_index(['date','econ']).sort_index()

if CFG['FLAGS_EXTRA']['ALT_CRED_R12_HAR']:
    for tag in ['R12_ALT','HAR_ALT']:
        colname = f'cred_alt_{tag}'
        if colname not in panel.columns:
            panel[colname] = np.nan
            for e in CFG['ECON']:
                c = f'cred_sc_{e}_{tag}'
                if c in panel.columns:
                    panel.loc[(slice(None), e), colname] = panel.loc[(slice(None), e), c]

        X_alt = ['US_for_EU','US_for_UK','CLI','dSR'] + [c for c in panel.columns if c.startswith('m_')] + [colname]
        res_alt = run_fe(panel.copy(), 'abs_y', X_alt, dk_bw=dk_BW, time_fe=False)
        print(f"\n=== FE with ALT credibility [{tag}] ===")
        print(res_alt.summary)

print("\n=== PANEL FE: abs_y on credibility (own + US interactions) + controls ===")
fe_abs = run_fe(panel.copy(), 'abs_y', baseline_X(panel, CFG), dk_bw=dk_BW, time_fe=False)
print(fe_abs.summary)

print("\n=== PANEL FE (robustness): new_y (signed) on same regressors ===")
fe_new = run_fe(panel.copy(), 'new_y', baseline_X(panel, CFG), dk_bw=dk_BW, time_fe=False)
print(fe_new.summary)

if CFG['FLAGS_EXTRA']['TIME_FE_MAIN']:
    print("\n=== MAIN PANEL FE (entity + time FE) ===")
    Xmain = [c for c in baseline_X(panel, CFG) if not c.startswith('m_')]
    fe_main = run_time_fe_baseline(panel.copy(), y='abs_y', xcols=Xmain, dk_bw=dk_BW)
    print(fe_main.summary)
else:
    fe_main = None

if CFG['FLAGS_EXTRA']['TRENDS_APPENDIX']:
    print("\n=== Appendix: country-specific trends ON ===")
    CFG['COUNTRY_TRENDS'] = True
    Xtrend = [c for c in baseline_X(panel, CFG) if not c.startswith('m_')]
    fe_trend = run_fe(panel.copy(), 'abs_y', Xtrend, dk_bw=dk_BW, time_fe=True)
    print(fe_trend.summary)
    CFG['COUNTRY_TRENDS'] = False
else:
    fe_trend = None

robust_rows = []
if CFG['FLAGS']['ROBUST']:
    for anchor_end in CFG['ECDF_ANCHOR_ENDS']:
        tmp = build_prob_panel(anchor_end, panel)
        for col in [c for c in panel.columns if (c.startswith('m_') or c in ['CLI','dSR','abs_y','new_y'])]:
            if col not in tmp.columns:
                tmp[col] = panel[col]
        resA = run_fe(tmp, 'abs_y', baseline_X(tmp, CFG), dk_bw=dk_BW, time_fe=False)
        robust_rows.append([anchor_end, float(resA.params.get('cred_sc_OWN', np.nan)),
                            float(resA.pvalues.get('cred_sc_OWN', np.nan)),
                            int(resA.nobs), float(resA.rsquared_within)])
robust_tab = pd.DataFrame(robust_rows, columns=['anchor_end','coef_own','pval_own','N','R2_within'])
if not robust_tab.empty:
    print("\n=== Anchor robustness (ECDF, pooled=True) ===")
    print(robust_tab)

def slice_period(df, start, end):
    return df.reset_index().query("date>=@start and date<=@end").set_index(['date','econ']).sort_index()

pre  = slice_period(panel, pd.Timestamp('2008-01-01'), pd.Timestamp('2013-12-01'))
post = slice_period(panel, pd.Timestamp('2014-01-01'), pd.Timestamp('2022-01-01'))

print("\n=== Subsamples FE (pre/post 2013-12) ===")
print("[PRE 2008–2013]")
res_pre  = run_fe(pre,  'abs_y', baseline_X(pre, CFG),  dk_bw=dk_BW, time_fe=False);  print(res_pre.summary)
print("[POST 2014–2022]")
res_post = run_fe(post, 'abs_y', baseline_X(post, CFG), dk_bw=dk_BW, time_fe=False);  print(res_post.summary)

panel_int = panel.reset_index().copy()
panel_int['post2013'] = (panel_int['date'] >= pd.Timestamp('2014-01-01')).astype(int)
panel_int['cred_sc_OWN_post'] = panel_int['cred_sc_OWN'] * panel_int['post2013']
Xint = baseline_X(panel_int, CFG) + ['cred_sc_OWN_post','post2013']
panel_int = panel_int.set_index(['date','econ']).sort_index()
print("\n=== Interaction FE: post2013 × credibility ===")
res_inter = run_fe(panel_int, 'abs_y', Xint, dk_bw=dk_BW, time_fe=False); print(res_inter.summary)

if CFG['FLAGS_EXTRA']['INTERACTIONS']:
    print("\n=== FE with interactions: cred×dSR & cred×SR ===")
    panel_m = panel.reset_index().copy()
    panel_m['cred_x_dSR'] = panel_m['cred_sc_OWN'] * panel_m['dSR']
    panel_m['cred_x_SR']  = panel_m['cred_sc_OWN'] * panel_m['SR']
    panel_m = panel_m.set_index(['date','econ']).sort_index()

    Xint2 = [c for c in baseline_X(panel_m, CFG) if not c.startswith('m_')] + ['cred_x_dSR','cred_x_SR']

    res_mech = run_fe(panel_m, 'abs_y', Xint2, dk_bw=dk_BW, time_fe=True)
    print(res_mech.summary)

placebo_rows = []
for L in [3,6,12]:
    tmp = panel.reset_index().sort_values(['econ','date']).copy()
    tmp[f'cred_sc_OWN_lead{L}'] = tmp.groupby('econ')['cred_sc_OWN'].shift(-L)
    base = baseline_X(tmp, CFG)
    base = [c for c in base if not (c.startswith('m_') or c.startswith('trend_'))]
    keep = base + [f'cred_sc_OWN_lead{L}','cred_sc_OWN']
    tmp = tmp.set_index(['date','econ']).sort_index()
    resL = run_fe(tmp, 'abs_y', keep, dk_bw=dk_BW, time_fe=True)
    placebo_rows.append([f'lead{L}',
                         float(resL.params.get(f'cred_sc_OWN_lead{L}', np.nan)),
                         float(resL.pvalues.get(f'cred_sc_OWN_lead{L}', np.nan)),
                         int(resL.nobs), float(resL.rsquared_within)])
placebo_tab = pd.DataFrame(placebo_rows, columns=['lead','coef','pval','N','R2_within'])
print("\n=== Lead-placebo FE (entity+time FE) ===")
print(placebo_tab)

dm_tab = None
if CFG['FLAGS']['FOREC']:
    print("\n=== Forecast tests: MSE & QLIKE + DM (HLN) + CPA (GW) ===")
    rows = []
    for econ in CFG['ECON']:
        df_e = panel.reset_index().set_index('date').query("econ==@econ").copy().sort_index()
        for H in CFG['H_LIST']:
            sigA = vol_sr_df[f'{econ}_EGARCH_t'].reindex(df_e.index)
            rv   = realized_vol(df_e['SR'], H, variance=False)
            sigR12 = df_e['SR'].shift(1).rolling(12, min_periods=12).std(ddof=0)
            sigHAR = har_mix(df_e['SR'])
            Z = make_instruments(df_e, 'SR', rv)
            for name, bench in [('EGARCH vs R12', sigR12), ('EGARCH vs HAR', sigHAR)]:
                d_mse = (rv - sigA)**2 - (rv - bench)**2
                t_dm, p_dm, t_hln, p_hln, T = dm_test_hln(d_mse, h=H, lags=H-1)
                F_cpa, p_cpa, t_u, p_u, _ = gw_test_conditional(d_mse, Z, lags=H-1)
                rows.append([econ,'SR',H,'MSE',name,T,t_dm,p_dm,t_hln,p_hln,_winner(name,t_dm),F_cpa,p_cpa,t_u,p_u])
                rv_var = realized_vol(df_e['SR'], H, variance=True)
                fA_var = (sigA**2).clip(lower=1e-10)
                fB_var = (bench**2).clip(lower=1e-10)
                d_ql = qlike_loss(rv_var, fA_var) - qlike_loss(rv_var, fB_var)
                t_dm, p_dm, t_hln, p_hln, T = dm_test_hln(d_ql, h=H, lags=H-1)
                F_cpa, p_cpa, t_u, p_u, _ = gw_test_conditional(d_ql, Z, lags=H-1)
                rows.append([econ,'SR',H,'QLIKE',name,T,t_dm,p_dm,t_hln,p_hln,_winner(name,t_dm),F_cpa,p_cpa,t_u,p_u])
        for H in CFG['H_LIST']:
            sigA = vol_inf_df[f'{econ}_EGARCH_t'].reindex(df_e.index)
            rv   = realized_vol(df_e['INF'], H, variance=False)
            sigR12 = df_e['INF'].shift(1).rolling(12, min_periods=12).std(ddof=0)
            sigHAR = har_mix(df_e['INF'])
            Z = make_instruments(df_e, 'INF', rv)
            for name, bench in [('EGARCH vs R12', sigR12), ('EGARCH vs HAR', sigHAR)]:
                d_mse = (rv - sigA)**2 - (rv - bench)**2
                t_dm, p_dm, t_hln, p_hln, T = dm_test_hln(d_mse, h=H, lags=H-1)
                F_cpa, p_cpa, t_u, p_u, _ = gw_test_conditional(d_mse, Z, lags=H-1)
                rows.append([econ,'INF',H,'MSE',name,T,t_dm,p_dm,t_hln,p_hln,_winner(name,t_dm),F_cpa,p_cpa,t_u,p_u])
                rv_var = realized_vol(df_e['INF'], H, variance=True)
                fA_var = (sigA**2).clip(lower=1e-10)
                fB_var = (bench**2).clip(lower=1e-10)
                d_ql = qlike_loss(rv_var, fA_var) - qlike_loss(rv_var, fB_var)
                t_dm, p_dm, t_hln, p_hln, T = dm_test_hln(d_ql, h=H, lags=H-1)
                F_cpa, p_cpa, t_u, p_u, _ = gw_test_conditional(d_ql, Z, lags=H-1)
                rows.append([econ,'INF',H,'QLIKE',name,T,t_dm,p_dm,t_hln,p_hln,_winner(name,t_dm),F_cpa,p_cpa,t_u,p_u])
    dm_tab = pd.DataFrame(
        rows,
        columns=['econ','var','H','loss','cmp','T','DM_t','DM_p','HLN_t','HLN_p',
                 'Winner_DM','GW_F','GW_p','GW_tα','GW_pα']
    )
    print(dm_tab.head())

oos_gap_tab = {}
if CFG['FLAGS_EXTRA']['MACRO_OOS']:
    print("\n=== OOS macro validation for |gap| (cred vs RW/AR1) ===")
    for econ in CFG['ECON']:
        df_e = (panel.reset_index().set_index('date').query("econ==@econ")
                [['abs_y','cred_sc_OWN','CLI','dSR','SR']].sort_index())
        oos = oos_gap_forecasts(df_e, H_list=CFG['H_LIST'], start_oos=pd.Timestamp('2012-01-01'))
        oos['econ']=econ
        oos_gap_tab[econ]=oos
        print(econ, '\n', oos)
    oos_gap_tab = pd.concat(oos_gap_tab, axis=0).reset_index(drop=True)

if CFG['FLAGS']['ECON']:
    print("\n=== BigMove leads (6m) – Entity FE + month dummies (DK) ===")
    df = panel.reset_index().sort_values(['econ','date']).copy()

    df['BigMove'] = (df.groupby('econ')['POLICY'].diff().abs() > CFG['BIGMOVE_THRESH']).astype(int)
    df['BigMove_lead6'] = df.groupby('econ')['BigMove'].transform(
        lambda x: x.shift(-1).rolling(6, min_periods=1).max()
    )

    feat_months = [c for c in df.columns if c.startswith('m_')]
    keep = ['BigMove_lead6','CLI','dSR','cred_sc_OWN','US_for_EU','US_for_UK'] + feat_months

    d2 = (df[keep + ['econ','date']]
            .dropna()
            .set_index(['econ','date'])
            .sort_index())

    y = d2['BigMove_lead6']
    X = d2.drop(columns=['BigMove_lead6'])
    X = X.loc[:, ~X.columns.duplicated(keep='first')]
    X = X.drop(columns=[c for c in X.columns if X[c].nunique() <= 1], errors='ignore')
    X, _ = drop_collinear_qr(X)
    y = y.loc[X.index]

    mod_bm = PanelOLS(y, X, entity_effects=True, time_effects=False)
    res_bm = mod_bm.fit(cov_type='kernel', kernel='bartlett', bandwidth=dk_BW)
    print(res_bm.summary)
else:
    res_bm = None

classif_tabs = {}
if CFG['FLAGS_EXTRA']['CLASSIFY']:
    print("\n=== BigMove classification (logit/probit) ===")
    df = panel.reset_index().sort_values(['econ','date']).copy()
    df['BigMove'] = (df.groupby('econ')['POLICY'].diff().abs() > CFG['BIGMOVE_THRESH']).astype(int)
    for H in [3,6,12]:
        df[f'BM_lead{H}'] = df.groupby('econ')['BigMove'].transform(lambda x: x.shift(-1).rolling(H, min_periods=1).max())
        feats = ['cred_sc_OWN','US_for_EU','US_for_UK','CLI','dSR'] + [c for c in df.columns if c.startswith('m_')]
        entD = pd.get_dummies(df['econ'], prefix='ent', drop_first=True, dtype=int)
        X = pd.concat([df[feats], entD], axis=1).dropna()
        y = df.loc[X.index, f'BM_lead{H}']
        Xc = sm.add_constant(X)
        logit = sm.Logit(y, Xc).fit(disp=False, maxiter=200)
        p_hat = logit.predict(Xc)
        auc = auc_from_scores(y.values, p_hat.values)
        cal = calibration_table(y.values, p_hat.values)
        classif_tabs[f'logit_H{H}'] = dict(model=logit, auc=auc, calib=cal)
        print(f'H={H} | AUC(logit)={auc:.3f} | N={int(y.shape[0])}')
        if CFG['FLAGS']['PLOTS'] and CFG['FLAGS_EXTRA']['CAMERA_READY']:
            plt.figure(figsize=(5,3))
            plt.plot(cal['pred'].values, cal['obs'].values, marker='o')
            plt.plot([0,1],[0,1], linestyle='--')
            plt.title(f'Calibration (H={H})'); plt.tight_layout(); plt.show()
        probit = sm.Probit(y, Xc).fit(disp=False, maxiter=200)
        p_hat2 = probit.predict(Xc)
        auc2 = auc_from_scores(y.values, p_hat2.values)
        classif_tabs[f'probit_H{H}'] = dict(model=probit, auc=auc2)

ev_outputs = {}; ev_meta = {}
if CFG['FLAGS']['EVENTS']:
    try:
        us_sr = panel.xs('US', level='econ')['SR'].dropna()
        idxs  = rpt.Binseg(model='l2').fit(us_sr.to_numpy()).predict(n_bkps=3)[:-1]
        dates = list(us_sr.index[idxs])
        print("US SR breakpoints:", [d.date() for d in dates])
        ev_meta['US_SR_breakpoints'] = [d.strftime('%Y-%m-%d') for d in dates]
        Wpre, Wpost = 6, 12
        def window_slice(s, t): return s.loc[t - pd.DateOffset(months=Wpre) : t + pd.DateOffset(months=Wpost)]
        for econ in CFG['ECON']:
            ccol = f'cred_sc_{econ}_EGARCH_t'; s = panel.xs(econ, level='econ')[ccol].dropna()
            mats = []
            for t in dates:
                w = window_slice(s, t)
                if len(w) == Wpre+Wpost+1:
                    w.index = range(-Wpre, Wpost+1); mats.append(w)
            if mats: ev_outputs[(econ,'event_mean')] = pd.concat(mats, axis=1).mean(axis=1)
        all_dates = panel.index.get_level_values('date').unique().sort_values()
        ok_mask = (all_dates >= all_dates.min() + pd.DateOffset(months=Wpre)) & \
                  (all_dates <= all_dates.max() - pd.DateOffset(months=Wpost+1))
        pool = all_dates[ok_mask]
        place_res = {}
        for econ in CFG['ECON']:
            s = panel.xs(econ, level='econ')[f'cred_sc_{econ}_EGARCH_t'].dropna()
            mats = []; draws = 0
            while draws < 19 and len(pool) >= len(dates):
                pick = np.random.choice(pool, size=len(dates), replace=False)
                tmp = []; ok = True
                for t in pick:
                    w = window_slice(s, t)
                    if len(w) != Wpre+Wpost+1: ok=False; break
                    w.index = range(-Wpre, Wpost+1); tmp.append(w)
                if ok and tmp:
                    mats.append(pd.concat(tmp, axis=1).mean(axis=1)); draws += 1
            if mats:
                place_res[econ] = pd.concat(mats, axis=1)
                ev_outputs[(econ,'placebo_mean')] = place_res[econ].mean(axis=1)
                ev_outputs[(econ,'placebo_p05')]  = place_res[econ].quantile(.05, axis=1)
                ev_outputs[(econ,'placebo_p95')]  = place_res[econ].quantile(.95, axis=1)
        if place_res:
            counts = {k: v.shape[1] for k,v in place_res.items()}
            print("Placebo draws used per econ (<= 500):", counts)
            ev_meta['placebo_draws'] = counts
    except Exception as e:
        logging.warning(f"Event study failed: {e}")

lp_tab = pd.DataFrame()
if CFG['FLAGS']['LP_VALIDATION']:
    rows = []
    for H in [6,12,24]:
        yH = panel.groupby('econ')['abs_y'].shift(-H)
        df_lp = (panel.copy()
                 .assign(abs_y_lead=yH)
                 .dropna(subset=['abs_y_lead','cred_sc_OWN'])
                 .reset_index()
                 .set_index(['econ','date'])
                 .sort_index())
        Xcols = [c for c in df_lp.columns if c.startswith('m_')] + ['cred_sc_OWN','CLI','dSR']
        Xcols = [c for c in Xcols if c in df_lp.columns]
        Xlp, _ = drop_collinear_qr(df_lp[Xcols])
        res = PanelOLS(df_lp.loc[Xlp.index, 'abs_y_lead'], Xlp,
                       entity_effects=True).fit(cov_type='kernel',
                                                kernel='bartlett',
                                                bandwidth=dk_BW)
        rows.append([H,
                     float(res.params.get('cred_sc_OWN', np.nan)),
                     float(res.pvalues.get('cred_sc_OWN', np.nan)),
                     int(res.nobs),
                     float(res.rsquared_within)])
    lp_tab = pd.DataFrame(rows, columns=['H','b_cred','pval','N','R2_within'])
    print("\n=== Predictive anchoring (LP on future |gap|) ===")
    print(lp_tab)

cred_raw_dc = pd.DataFrame(); cred_sc_dc = pd.DataFrame()
vol_sr_gjr = pd.DataFrame(); cred_sc_gjr = pd.DataFrame()
res_iv1 = None; res_iv2 = None; plc = pd.DataFrame(); dyn = pd.DataFrame()
dm_oos = pd.DataFrame(); mcs_tab = pd.DataFrame()

try:
    sigma_decom = pca_decommonize(vol_sr_df.copy())
    cred_raw_dc , cred_sc_dc , _ = build_scaled_from_sigma(sigma_decom, CFG['ANCHOR'], prefix='DECOM')
    _dc = cred_sc_dc.rename(columns={c:f'cred_sc_{c.split("_")[2]}_DECOM' if c.startswith('cred_sc_') else c for c in cred_sc_dc.columns})
    panel = (panel.reset_index().merge(_dc.reset_index(), on='date', how='left')
                      .set_index(['date','econ']).sort_index())
    panel['cred_dc_OWN'] = np.nan
    for e in CFG['ECON']:
        col = f'cred_sc_{e}_DECOM'
        if col in panel.columns:
            panel.loc[(slice(None), e), 'cred_dc_OWN'] = panel.loc[(slice(None), e), col]
except Exception as e:
    logging.warning(f"De-commonize failed: {e}")

try:
    vol_cols_sr_gjr = {}
    for econ, g in panel['SR'].groupby('econ'):
        y = g.droplevel('econ').dropna()
        p = select_ar(y, CFG['AR_MAX'])
        _, sigG = fit_gjr_garch(y, p)
        vol_cols_sr_gjr[econ] = sigG
    vol_sr_gjr = pd.DataFrame(vol_cols_sr_gjr)
    _, cred_sc_gjr, _ = build_scaled_from_sigma(vol_sr_gjr, CFG['ANCHOR'], prefix='GJR')
    _corr = {}
    for e in CFG['ECON']:
        a = panel.xs(e, level='econ')[f'cred_sc_{e}_EGARCH_t']
        b = cred_sc_gjr[f'cred_sc_{e}_GJR']
        _corr[e] = a.corr(b)
    print("Alt-model (GJR) vs EGARCH – correlations:", _corr)
except Exception as e:
    logging.warning(f"GJR alt failed: {e}")

try:
    print("\n=== IV Panel (manual 2SLS with entity FE) ===")
    res_iv1, res_iv2 = run_iv_panel_manual(panel, dk_bw=dk_BW)
    print("[1st stage] cred_sc_OWN ~ IVs + exog")
    print(res_iv1.summary)
    print("[2nd stage] abs_y ~ cred_hat + exog")
    print(res_iv2.summary)
except Exception as e:
    logging.warning(f"IV panel failed: {e}")

try:
    print("\n=== Country-swap placebo (US broadcast shuffled over time) ===")
    plc = country_swap_placebo(panel, K=100, dk_bw=dk_BW)
    print(plc.describe())
except Exception as e:
    logging.warning(f"Placebo swap failed: {e}")

plc_brd = pd.DataFrame()
if CFG['FLAGS_EXTRA']['TARGETED_PLACEBO']:
    print("\n=== Targeted placebo: US_for_EU/UK under shuffled US broadcast ===")
    plc_brd = country_swap_placebo_broadcast(panel, K=200, dk_bw=dk_BW)
    print(plc_brd.describe())

try:
    dyn = build_dynamic_ecdf_prob(vol_sr_df, CFG['ECDF_ANCHOR_ENDS'])
    us_cols = [c for c in dyn.columns if c[1]=='US']
    if us_cols:
        tmp = dyn[us_cols].dropna()
        print("\n=== Dynamic ECDF – pairwise corr (US) ===")
        print(tmp.corr())
except Exception as e:
    logging.warning(f"Dynamic ECDF failed: {e}")

try:
    dm_oos, mcs_tab = fair_horserace_and_mcs(panel, vol_sr_df, CFG['H_LIST'], start_oos='2012-01-01')
    print("\n=== Real-time OOS DM/CPA (vs R12 baseline) ===")
    print(dm_oos.head(12))
    print("\n=== MCS survivors (α=10%) ===")
    print(mcs_tab.head())
except Exception as e:
    logging.warning(f"Horse-race/MCS failed: {e}")

if CFG['FLAGS']['PLOTS']:
    for econ in CFG['ECON']:
        base = vol_sr_df[f'{econ}_EGARCH_t'].dropna()
        plt.figure(figsize=(10,3.6)); base.plot(label=f'σ̂ {econ} EGARCH-t', lw=2)
        plt.title(f'{econ}: EGARCH-t σ̂ (SR)'); plt.xlabel('date'); plt.legend(); plt.tight_layout(); plt.show()
    for econ in CFG['ECON']:
        col = f'cred_sc_{econ}_EGARCH_t'
        if col in panel.columns:
            plt.figure(figsize=(7,3)); panel.xs(econ, level='econ')[col].plot()
            plt.title(f'{econ}: Credibility (scaled 0–100, EGARCH-t)'); plt.grid(alpha=.3); plt.tight_layout(); plt.show()

bands_dict = {}
if CFG['FLAGS_EXTRA']['BANDS']:
    print("\n=== Bands (MBB) for credibility indices ===")
    for econ in CFG['ECON']:
        col = f'cred_sc_{econ}_EGARCH_t'
        if col in panel.columns:
            s = panel.xs(econ, level='econ')[col].dropna()
            b = cred_bands_mbb(s, block=CFG['MBLOCK']['block'], reps=CFG['MBLOCK']['reps'])
            bands_dict[econ] = b

            if CFG['FLAGS']['PLOTS'] and CFG['FLAGS_EXTRA']['CAMERA_READY']:
                plt.figure(figsize=(7,3))
                s.plot(label='cred')
                b['low'].plot(linestyle='--', alpha=.7)
                b['high'].plot(linestyle='--', alpha=.7)
                plt.title(f'{econ}: Credibility (0–100) with MBB bands')
                plt.tight_layout(); plt.show()

out = pathlib.Path('/content/credibility_planB.xlsx')
with pd.ExcelWriter(out) as w:
    vol_sr_ser.to_frame('sigma_SR').to_excel(w, sheet_name='vol_SR')
    vol_inf_df.to_excel(w, sheet_name='vol_INF')
    cred_raw.to_excel(w, sheet_name='cred_raw(-logσ_SR)')
    scaled_all.to_excel(w, sheet_name='cred_scaled_main')

    tmp_exp = panel[['INF','exp_INF','target','new_y','abs_y']].reset_index()
    tmp_exp.to_excel(w, sheet_name='expectations_proxy', index=False)

    pd.DataFrame({'FE_abs_y':[fe_abs.summary.as_text()]}).to_excel(w, sheet_name='FE_abs_y_txt', index=False)
    pd.DataFrame({'FE_new_y':[fe_new.summary.as_text()]}).to_excel(w, sheet_name='FE_new_y_txt', index=False)
    if not robust_tab.empty:
        robust_tab.to_excel(w, sheet_name='FE_anchor_ECDF', index=False)
    pd.DataFrame({'FE_pre_txt':[res_pre.summary.as_text()]}).to_excel(w, sheet_name='FE_pre_txt', index=False)
    pd.DataFrame({'FE_post_txt':[res_post.summary.as_text()]}).to_excel(w, sheet_name='FE_post_txt', index=False)
    pd.DataFrame({'FE_interact_txt':[res_inter.summary.as_text()]}).to_excel(w, sheet_name='FE_interact_txt', index=False)
    placebo_tab.to_excel(w, sheet_name='FE_placebo_leads', index=False)

    if dm_tab is not None and not dm_tab.empty:
        dm_tab.to_excel(w, sheet_name='Forecast_tests', index=False)

    if not lp_tab.empty:
        lp_tab.to_excel(w, sheet_name='LP_predictive', index=False)

    if 'res_bm' in locals() and res_bm is not None:
        pd.DataFrame({'BigMove_FE':[res_bm.summary.as_text()]}).to_excel(w, sheet_name='bigmove_FE_txt', index=False)

    if ev_outputs:
        pd.concat(ev_outputs, axis=1).to_excel(w, sheet_name='event_study')
        meta_df = pd.DataFrame({'key':['US_SR_breakpoints','placebo_draws'],
                                'value':[json.dumps(ev_meta.get('US_SR_breakpoints', [])),
                                         json.dumps(ev_meta.get('placebo_draws', {}))]})
        meta_df.to_excel(w, sheet_name='event_meta', index=False)

    if not cred_raw_dc.empty:
        cred_raw_dc.to_excel(w, sheet_name='cred_raw_decommon', index=True)
    if not cred_sc_dc.empty:
        cred_sc_dc.to_excel(w, sheet_name='cred_scaled_decommon', index=True)
    if not vol_sr_gjr.empty:
        vol_sr_gjr.to_excel(w, sheet_name='vol_SR_GJR', index=True)
    if not cred_sc_gjr.empty:
        cred_sc_gjr.to_excel(w, sheet_name='cred_scaled_GJR', index=True)
    if res_iv1 is not None and res_iv2 is not None:
        pd.DataFrame({'IV_first_stage':[res_iv1.summary.as_text()],
                      'IV_second_stage':[res_iv2.summary.as_text()]}).to_excel(w, sheet_name='IV_panel_txt', index=False)
    if not plc.empty:
        plc.to_excel(w, sheet_name='placebo_swap_beta', index=False)
    if not dm_oos.empty:
        dm_oos.to_excel(w, sheet_name='Forecast_OOS_DMCPA', index=False)
    if not mcs_tab.empty:
        mcs_tab.to_excel(w, sheet_name='Forecast_MCS', index=False)
    if not dyn.empty:
        dyn.to_excel(w, sheet_name='ECDF_dynamic', index=True)
    if 'fe_main' in locals() and fe_main is not None:
        pd.DataFrame({'FE_main_timeFE_txt':[fe_main.summary.as_text()]}).to_excel(w, sheet_name='FE_main_timeFE', index=False)
    if 'fe_trend' in locals() and fe_trend is not None:
        pd.DataFrame({'FE_trends_txt':[fe_trend.summary.as_text()]}).to_excel(w, sheet_name='FE_trends_appendix', index=False)

    if 'exp_grid_tabs' in locals() and exp_grid_tabs:
        for econ, tab in exp_grid_tabs.items():
            tab.to_excel(w, sheet_name=f'EXP_grid_{econ}')

    if 'cred_alt_df' in locals():
        cred_alt_df.to_excel(w, sheet_name='cred_ALT_R12_HAR', index=False)

    if 'bands_dict' in locals() and bands_dict:
        for econ,b in bands_dict.items():
            b.to_excel(w, sheet_name=f'bands_{econ}')

    if 'oos_gap_tab' in locals() and isinstance(oos_gap_tab, pd.DataFrame):
        oos_gap_tab.to_excel(w, sheet_name='OOS_gap_DMCPA', index=False)

    if 'classif_tabs' in locals() and classif_tabs:
        summ = []
        for k,v in classif_tabs.items():
            if 'auc' in v: summ.append([k, v['auc']])
        pd.DataFrame(summ, columns=['model','AUC']).to_excel(w, sheet_name='BigMove_AUC', index=False)
        for k,v in classif_tabs.items():
            if 'calib' in v:
                v['calib'].to_excel(w, sheet_name=f'Calib_{k}', index=False)

    if 'plc_brd' in locals() and not plc_brd.empty:
        plc_brd.to_excel(w, sheet_name='Placebo_USbroadcast', index=False)

    pd.DataFrame([{'anchor_scaled':'pre2010', **anchor_stats}]).to_excel(w, sheet_name='anchor_stats', index=False)
    pd.DataFrame({'config':[json.dumps(CFG, default=str)]}).to_excel(w, sheet_name='config_json', index=False)

print("✅ Saved", out)
```
